```python
# 조건문 표현식 (Conditional Expression) if ~else문을 한줄에 작성
score = 85
result = "Success" if score >= 80 else "Fail" #참값이 왼쪽, 거짓이 오른쪽 
print(result)
```

# 반복문
- while보다 for문이 코드가 더 간결한 경우가 많음

```python
i = 1
result = 0
while i<=9:
    result += i 
    i+=1
print(result)
```

```python
# for 변수 in 리스트: 
# in 뒤에 오는 데이터(리스트, 튜플 등)에 포함되어 있는 원소를 첫번째 인덱스 부터 차례대로 하나씩 방문

array = [9,8,7,6,5]
for x in array:
    print(x)
```

```python
array = (1,2,3,4,5)
for x in array:
    print(x)
```

```python
# range(시작 값, 끝 값+1) 형태로 사용, 인자를 하나만 넣으면 시작값은 0
result = 0
for i in range(1,10):
    result += i
print(result)
```

```python
# continue: 반복문에서 남은 코드의 실행을 건너뛰고, 다음 반복을 진행하고자 할 때 continue
# 1부터 9까지의 홀수의 합을 구할 때 
result = 0
for i in range(10):
    if i % 2 ==0:
        continue # 짝수는 일단 넘김
    result += i
print(result)
        
```

```python
# 반복문 즉시탈출 break 
i =1 
while True:
    print("현재 i의 값", i)
    if i == 5:
        break
    i += 1
```

```python
scores = [90, 85, 77, 65, 97 ]
for i in range(5):
    if scores[i] >= 80:
        print(f"합격입니다 {i+1}번 학생~")
```

```python
# 특정 번호의 학생은 제외하기
scores = [90, 85, 77, 65, 97]
cheating_student_list = {2,4}
for i in range(5):
    if i+1 in cheating_student_list:
        continue
    if scores[i] >= 80:
        print(f"합격입니다 {i+1}번 학생~")
        
        
```

# 함수와 람다표현식
- 함수 매개변수: 함수 내부에서 사용할 변수
- 반환값: 함수에서 처리된 결과를 반환 
- def 함수명(매개변수)
 * 실행할 소스코드
 * return 반환 값

```python
# 파라미터의 변수를 직접 지정 가능, 매개변수의 순서가 달라도 상관 없음
def add(a,b):
    print('함수의 결과:', a+b)

add(b=3, a=7)
```

```python
# global 키워드: 함수 바깥에 선언된 변수를 바로 참조 
a = 0
def func():
    global a
    a+=1
    
for i in range(10):
    func()
print(a)
    
```

```python
# 람다표현식을 이용하면 특정한 기능을 수행하는 함수를 한 줄에 작성할 수 있음 

print((lambda a, b: a+b)(3,7)) # (lambda 입력 매개변수 : 소스코드)(파라미터)
```

```python
# 내장함수에서 자주 사용됨

array = [('홍길동', 50), ('이순신', 32), ('아무개', 74)]

def my_key(x):
    return x[1]

print(sorted(array, key=my_key)) # 점수순 오름차순 정렬 
print(sorted(array, key=lambda x: x[1]))
```

```python
# 람다 표현식 예시: 여러 개의 리스트에 적용
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
result = map(lambda a, b: a+b,list1, list2) # map: 각각의 원소에 어떠한 함수 적용 
print(list(result))
```

# 실전에서 유용한 표준 라이브러리


```python
# 자주 사용되는 내장함수 
result = sum([1,2,3,4,5])

min_result = min(7,3,5,2)
max_result = max(7,3,5,2)

result = eval("(3+5)*7")
print(result)
```

```python
# sorted(): 각 원소를 정렬한 결과 반환

result = sorted([9,1,8,5,4])
reverse_result = sorted([9,1,8,5,4], reverse=True)
print(result)
print(reverse_result)

# sorted() with key 
array = [('홍길동', 50), ('이순신', 32), ('아무개', 74)]
result = sorted(array, key=lambda x: x[1], reverse=True) # 현재원소에서 두번째 원소를 기준으로함
print(result)
```

```python
# 순열
from itertools import permutations
data = ['A', 'B', 'C']
result = list(permutations(data,3)) # 모든 순열 구하기 3P3
print(result)
```

```python
# 조합 
from itertools import combinations
from itertools import permutations
data = ['A', 'B', 'C']
result = list(combinations(data,2)) # 조합구하기 3C2
print(result)
```

```python
# Counter: 등장 횟수 
# 리스트와 같은 반복 가능한 객체가 주어졌을 때 내부의 원소가 몇 번씩 등장했는지 알려줌
from collections import Counter
counter = Counter(['red', 'blue', 'red', 'green', 'blue', 'blue'])
print(counter['blue'])
print(counter['green'])
print(dict(counter))
```

```python
#최대공약수 GCD , 최소공배수 LCM

import math 
a=21
b=14
print(math.gcd(21,14))

def lcm(a,b):
    return a*b // math.gcd(a,b)

print(lcm(21,14))
```
