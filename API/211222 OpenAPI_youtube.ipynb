## 한국환경공단_측정소정보
- https://www.data.go.kr/data/15073861/openapi.do

```python
import requests
import pandas as pd
from xml.etree import ElementTree 

skey = 'G2CKapwObthJV7ztImK+I43MXyqP7PqJVzkLXodcz2WjpaQfxBPbgueeWM6v2vIU7BJkpjLEbW+7mtCPRfejSw==' # decode key
url = 'http://apis.data.go.kr/B552584/MsrstnInfoInqireSvc/getNearbyMsrstnList'
params ={'serviceKey' : skey, 'returnType' : 'xml', 'tmX' : '244148.546388 ', 'tmY' : '412423.75772', 'ver': '1.0'}

response = requests.get(url, params=params)
response_body=response.text
print(response_body)
```

```python
root = ElementTree.fromstring(response_body)

df = pd.DataFrame()
for item in root.iter('item'):
    item_dict = {}
    item_dict['stationName'] = item.find('stationName').text
    item_dict['addr'] = item.find('addr').text
    item_dict['tm'] = item.find('tm').text
    df= df.append(item_dict, ignore_index=True)  
df
```
